## 2. Your First Agent

````
from yom import Agent

# Create a simple agent
agent = Agent(
    system_prompt="You are a friendly assistant. Keep answers short and fun!"
)

# Run synchronously
result = agent.run_sync("What is Python?")
print(result)
````

## 3. Conversation Memory

The agent remembers your conversation!

````
# First, tell the agent your name
agent.run_sync("My name is Alice")

# Ask about your name - it remembers!
result = agent.run_sync("What's my name?")
print(result)
````

## 4. Async Usage

````
import asyncio

async def chat():
    agent = Agent(system_prompt="You are helpful.")
    result = await agent.run("What is 2+2?")
    print(result)

asyncio.run(chat())
````

## 5. Available Models

yom auto-detects providers from model names:

````
# MiniMax models (uses MINIMAX_API_KEY)
agent = Agent(model="MiniMax-M2.7")
result = agent.run_sync("Say hello in 5 words")
print(result)
````

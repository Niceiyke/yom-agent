## 2. Built-in Tools

Tools extend what the agent can do. The `core` tools include file operations!

````
from yom import Agent

# Core tools: read, write, edit, bash, glob, grep
agent = Agent(tools=["core"])

# Use bash tool to run commands
result = agent.run_sync("""
Run: echo 'Hello from shell!' > /tmp/test.txt && cat /tmp/test.txt
""")
print(result)
````

## 3. HTTP Tool

The agent can make HTTP requests!

````
agent = Agent(tools=["http_request"])

result = agent.run_sync("""
Make a GET request to https://httpbin.org/get
Return just the response status code
""")
print(result)
````

## 4. Custom Tools

Create your own tools with the `@tool` decorator!

````
from yom import tool

# Define a custom tool
@tool(name="calculator", description="Perform basic math")
def calculate(a: float, b: float, operation: str) -> str:
    """Calculator tool."""
    if operation == "add":
        return str(a + b)
    elif operation == "subtract":
        return str(a - b)
    elif operation == "multiply":
        return str(a * b)
    elif operation == "divide":
        return str(a / b) if b != 0 else "Error: division by zero"
    return "Unknown operation"

# Use it
agent = Agent(tools=["core", calculate])
result = agent.run_sync("Calculate 10 + 5 using the calculator")
print(result)
````

## 5. Sessions

Each user gets their own conversation memory!

````
# User Alice's session
alice = Agent(session_id="alice", system_prompt="You are Alice's assistant.")
alice.run_sync("I love pizza")

# User Bob's session
bob = Agent(session_id="bob", system_prompt="You are Bob's assistant.")
bob.run_sync("I love pasta")

# Each remembers their own preference!
print("Alice:", alice.run_sync("What food do I love?"))
print("Bob:", bob.run_sync("What food do I love?"))
````

## 6. Testing Your Agent

Use fake agents for testing without API calls!

````
from yom.testing import fake_agent, assert_response

# Create a fake agent for testing
test_agent = fake_agent("I am a helpful assistant.")

# Run test
result = test_agent.run_sync("Who are you?")
print("Response:", result)

# Assert response contains expected text
assert_response(result, contains="helpful")
print("✅ Test passed!")
````
